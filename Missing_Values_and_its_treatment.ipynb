```python
import pandas as pd
import numpy as np
```

```python
data = {
    "A": [1,2,100,4,5],
    "B": [3,np.nan,7,8,9],
    "C": [np.nan,12,13,14,np.nan],
    "D": [16,17,18,np.nan,20]
}
```

```python
# convert the dictionary to a pandas DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
#dropping rows with null values
df.dropna()
```

```python
#dropping columns with null values
df.dropna(axis=1)
```

```python
sns.histplot(df.age, kde=True)
```

```python
# whenever you have normal distrubution data and no outlier , impute with mean

df["mean_imputation"] = df["age"].fillna(df["age"].mean())
```

```python
-df[["age", "mean_imputation"]]
```

```python
sns.distplot(df["mean_imputation"], kde = True)
```

```python
# mean imputation works well with normally distributed data / no outlier is persent
```

```python
df["median_imputation"] = df["age"].fillna(df["age"].median())
```

```python
df[["age", "mean_imputation","median_imputation"]]
```

```python
sns.distplot(df["median_imputation"], kde = True)
```
